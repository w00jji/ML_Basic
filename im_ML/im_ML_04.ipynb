### iris 데이터(랜덤 포레스트, 의사결정나무 성능 비교)

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. iris 데이터셋 로드 및 필터링
iris_df = sns.load_dataset('iris')
iris_df = iris_df[(iris_df['species'] == 'versicolor') | (iris_df['species'] == 'virginica')]

# 2. 특징 및 라벨 설정
X = iris_df[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']]  # 모든 특징 사용
Y = iris_df['species'].map({'versicolor': 0, 'virginica': 1})  # 라벨 인코딩

# 3. 훈련 세트와 테스트 세트로 분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)

# 4. 의사결정 트리 모델 정의 및 훈련
DT_Model = DecisionTreeClassifier(random_state=0)
DT_Model.fit(X_train, Y_train)

# 5. 랜덤 포레스트 모델 정의 및 훈련
RF_Model = RandomForestClassifier(random_state=0)
RF_Model.fit(X_train, Y_train)

# 6. 훈련 세트 및 테스트 세트에 대한 성능 비교
DT_Train_score = DT_Model.score(X_train, Y_train)
DT_Test_score = DT_Model.score(X_test, Y_test)

RF_Train_score = RF_Model.score(X_train, Y_train)
RF_Test_score = RF_Model.score(X_test, Y_test)

print('Decision Tree - Train Score:', DT_Train_score, 'Test Score:', DT_Test_score)
print('Random Forest - Train Score:', RF_Train_score, 'Test Score:', RF_Test_score)

# 7. 새 데이터를 사용한 예측
Pre_Parm = [[5.0, 1.5, 3.5, 1.0]]  # 예시 입력값
predict_DT = DT_Model.predict(Pre_Parm)
predict_RF = RF_Model.predict(Pre_Parm)

print('--------------')
print('Decision Tree Prediction:', predict_DT)
print('Random Forest Prediction:', predict_RF)

# 8. CSV 파일을 사용한 예측
Pre_DataSet = pd.read_csv('ml_sample/Prediction_Scenario.csv')

# CSV 파일의 열이 2개라면, 학습된 모델에 맞게 확장 (예시로 두 열을 중복 사용)
# 두 개의 열을 가지고 있는 CSV 파일을 예시로 사용하여 기존 열을 반복하여 학습 피처에 맞게 조정
Pre_DataSet.columns = ['petal_length', 'petal_width']  # 현재 존재하는 두 열
Pre_DataSet['sepal_length'] = Pre_DataSet['petal_length']  # 새로운 열 추가
Pre_DataSet['sepal_width'] = Pre_DataSet['petal_width']  # 새로운 열 추가

# 이제 변환된 데이터를 사용하여 예측 수행
predict_DT_sR = DT_Model.predict(Pre_DataSet)
predict_RF_sR = RF_Model.predict(Pre_DataSet)

print('--------------')
print('Decision Tree Predictions from CSV:', predict_DT_sR)
print('Random Forest Predictions from CSV:', predict_RF_sR)
```

### 의사결정나무 과적합 해결

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 데이터 로드 및 분할
iris_df = sns.load_dataset('iris')
iris_df = iris_df[(iris_df['species'] == 'versicolor') | (iris_df['species'] == 'virginica')]
X = iris_df[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']]
Y = iris_df['species'].map({'versicolor': 0, 'virginica': 1})
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)

# Decision Tree 모델 정의 및 하이퍼파라미터 설정
DT_Model = DecisionTreeClassifier(max_depth=3, min_samples_split=4, min_samples_leaf=2, random_state=0)
DT_Model.fit(X_train, Y_train)

# 모델 성능 평가
DT_Test_score = DT_Model.score(X_test, Y_test)
print('Decision Tree - Test Score:', DT_Test_score)
```

### 랜덤포레스트 과적합 해결

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 데이터 로드 및 분할
iris_df = sns.load_dataset('iris')
iris_df = iris_df[(iris_df['species'] == 'versicolor') | (iris_df['species'] == 'virginica')]
X = iris_df[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']]
Y = iris_df['species'].map({'versicolor': 0, 'virginica': 1})
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=0)

# Random Forest 모델 정의 및 하이퍼파라미터 설정
RF_Model = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=4, min_samples_leaf=2, random_state=0)
RF_Model.fit(X_train, Y_train)

# 모델 성능 평가
RF_Test_score = RF_Model.score(X_test, Y_test)
print('Random Forest - Test Score:', RF_Test_score)
```

```python
# 교차검증 - 홀드아웃 으로 해결
from sklearn.model_selection import cross_val_score

# 교차 검증을 통한 모델 성능 평가
DT_scores = cross_val_score(DT_Model, X, Y, cv=5)
RF_scores = cross_val_score(RF_Model, X, Y, cv=5)

print('Decision Tree - Cross-Validation Scores:', DT_scores)
print('Random Forest - Cross-Validation Scores:', RF_scores)
```

```python
# 시계열 데이터(날짜)를 lineplot을 통해 알아보기

plt.figure(figsize=(14,4))  # 그래프 크기 조정
sns.lineplot(x='date', y='count', data=bike_df)  
plt.xticks(rotation=45)  # x축 데이터를 45도 회전시켜서 보여줌
plt.show()

# 날짜 별 대여 개수
# 2020년1월에서 2020년 7월 중간에 이상한 부분이 있음
```

```python
# 해당 데이터의 상태를 비교해보기 위해 2019년도 데이터를 월별로 그룹하여 자전거 대여수 평균을 구하기

bike_df[bike_df['year'] == 2019].groupby('month')['count'].mean()
```

```python
# 2020년도 데이터를 월별로 그룹하여 자전거 대여수 평균 구하기

bike_df[bike_df['year'] == 2020].groupby('month')['count'].mean()
```

```python
# bike_df에 'covid'라는 새로운 파생변수를 만듦

# 날짜를 계절과 같은 범주형 데이터로 바꿈(# 2020-04-01 이전: precovid
# 2021-04-01 이전 : covid
# 그 이후 : postcovid

# 날짜를 계절과 같은 범주형 데이터로 바꾸는 함수 정의
def covid(date):
    if date < pd.Timestamp('2020-04-01'):
        return 'precovid'
    elif date < pd.Timestamp('2021-04-01'):
        return 'covid'
    else:
        return 'postcovid'

# 'date' 열을 datetime으로 변환
bike_df['date'] = pd.to_datetime(bike_df['date'])

# 데이터프레임에 함수 적용
bike_df['covid'] = bike_df['date'].apply(covid)
```

```python

bike_df.tail()
```

```python
# 'month' 컬럼으로 'season'이라는 새로운 파생변수를 만들기

# 3~5월 : spring
# 6~8 : summer
# 9~11 : fall
# 12~2 : winter

bike_df['season'] = bike_df['month'].apply(lambda month: 'winter' if month == 12 else 'fall' if month >= 9 else 'summer' if month >=6 else 'spring' if month >=3 else 'winter')
bike_df.head()
```

```python
# 'hour'컬럼으로 'day_night'라는 새로운 파생변수 만들기

bike_df['day_night'] = bike_df['hour'].apply(lambda x: 'night' if x>=21 else 'late evening' if x>=19 else 'early evening' if x>=17 else 'late afternoon' if x>=16 else 'early afternoon' if x>= 13 else 'late morning' if x>=11 else 'early morning' if x>=5 else 'night')
bike_df['day_night']
```

```python
# 모델 학습 시 필요없는 컬럼 삭제하고 저장하기

bike_df.drop(['datetime', 'month', 'date', 'hour'], axis=1, inplace=True)
bike_df.head()
```

#### 의사결정 나무
- 데이터를 분석하여 그 사이에 존재하는 패턴을 예측 가능한 규칙들의 조합으로 나타내며 그 모양이 나무와 같다고 하여 의사결정나무 라고 부름

- 회귀(Regression), 분류(Classification) 모두 가능
- 지니계수(Gini Index): 0에 가까울수록 클래스에 속한 불순도가 낮음
- 엔트로피(Entropy): 결정을 내릴만한 충분한 정보가 데이터에 없다고 보는 것. (0에 가까울수록 결정을 내릴만한 충분한 정보가 있다) 📍 지니계수와 엔트로피는 반비례 관계
-  오버피팅(과적합): 훈련데이터에서는 정확하나 테스트데이터에서는 성과가 나쁜 현상을 말함. 훈련 데이터가 적거나 노이즈가 있을 때 또는 알고리즘 자체가 나쁠 때 발생. 의사 결정 나무에서는 나무의 가지가 너무 많거나 크기가 클 때 발생

- 의사 결정 나무에서 오버피팅을 피하는 방법
  - 사전 가지치기: 나무가 다 자라기 전에 알고리즘을 멈추는 방법
  - 사후 가지치기: 의사 결정 나무를 끝까지 돌린 후 밑에서부터 가지를 쳐나가는 방법

### adaboost 과정 시각화 예제

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# 가상의 데이터 생성
X, y = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=0)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)

# AdaBoost 모델 훈련
model = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50, random_state=0)
model.fit(X_train, y_train)

def plot_decision_boundaries(X, y, model, ax, title):
    # 결정 경계 그리기
    h = .02  # 스텝 사이즈
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.8)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_title(title)

# 서브플롯 생성
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# 전체 AdaBoost 모델의 결정 경계
plot_decision_boundaries(X_train, y_train, model, axs[0], 'AdaBoost Model')

# 개별 약한 학습기의 결정 경계
for i, estimator in enumerate(model.estimators_):
    plot_decision_boundaries(X_train, y_train, estimator, axs[1], f'Weak Learner {i + 1}')

plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

# 데이터 불러오기 및 DataFrame 생성
iris = load_iris()
df1 = pd.DataFrame(iris.data, columns=iris.feature_names)
df1['species'] = [iris.target_names[x] for x in iris.target]

# 특징과 라벨 설정
X = df1.drop('species', axis=1)
y = df1['species']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)

# Gradient Boosting 모델 훈련
GBM_Model = GradientBoostingClassifier(random_state=0)
GBM_Model.fit(X_train, y_train)

# 예측 수행
y_pred = GBM_Model.predict(X_test)

# 모델 평가
Train_score = GBM_Model.score(X_train, y_train)
Test_score = GBM_Model.score(X_test, y_test)

# F1 스코어 계산
f1 = f1_score(y_test, y_pred, average='weighted')

print('GBM - Train_score:', Train_score, 'Test_Score:', Test_score)
print('GBM - F1 Score:', f1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder

# 타이타닉 데이터 불러오기 (df_cleaned는 이전에 처리된 데이터프레임이라고 가정)
# df_cleaned는 이전 단계에서 Age, Sex의 결측치를 제거하고 Cabin_Exists 열을 추가한 데이터프레임

# 특징과 라벨 설정
X = df_cleaned[['Age', 'Sex']]
y = df_cleaned['Survived']

# 성별 인코딩
label_encoder = LabelEncoder()
X['Sex'] = label_encoder.fit_transform(X['Sex'])

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)

# Gradient Boosting 모델 훈련
GBM_Model = GradientBoostingClassifier(random_state=0)
GBM_Model.fit(X_train, y_train)

# 예측 수행
y_pred = GBM_Model.predict(X_test)

# 모델 평가
Train_score = GBM_Model.score(X_train, y_train)
Test_score = GBM_Model.score(X_test, y_test)

# F1 스코어 계산
f1 = f1_score(y_test, y_pred, average='weighted')

print('GBM - Train_score:', Train_score)
print('GBM - Test_Score:', Test_score)
print('GBM - F1 Score:', f1)
```

```python
# 학습 과정에서의 손실 값 시각화
plt.figure(figsize=(10, 6))
plt.plot(GBM_Model.train_score_, label='Training Loss')
plt.title('Gradient Boosting Training Loss Over Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 학습 과정에서의 손실 값 시각화
train_loss = GBM_Model.train_score_

# 손실 변화율 계산 (기울기)
loss_diff = np.diff(train_loss)
min_slope_idx = np.argmin(np.abs(loss_diff))

# 시각화 시작
plt.figure(figsize=(14, 8))

# 첫 번째 그래프: 손실 곡선
plt.subplot(2, 1, 1)
plt.plot(train_loss, label='Training Loss', color='blue')
plt.axvline(min_slope_idx, color='red', linestyle='--', label='Slope Change Point')
plt.title('Gradient Boosting Training Loss Over Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Loss')
plt.legend()

# 두 번째 그래프: 손실 변화율 (기울기)
plt.subplot(2, 1, 2)
plt.plot(loss_diff, label='Loss Gradient', color='green')
plt.axvline(min_slope_idx, color='red', linestyle='--', label='Slope Change Point')
plt.title('Gradient of Loss Over Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
