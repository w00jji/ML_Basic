### 회귀분석 코드1

```python
import pandas as pd
import sklearn.model_selection as sms

# 데이터 읽어오기
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

pd_DataSet = pd.read_csv('ml_sample/한국주택금융공사_주택금융신용보증 현황_20190630.csv', encoding='euc-kr')

DF = pd.DataFrame(pd_DataSet)

# 결측값이 있는 행 삭제
New_DF = DF.dropna(how='any')

# 독립 변수와 종속 변수 설정
x = New_DF[['연도', '월', '구입 세대수(호)', '전세 세대수(호)', '전세 금액 ', '중도금 세대수(호)',
       '중도금 금액 ', '개량 금액 ', '사업자보증세대수(호)', '사업자보증금액 ']]
y = New_DF['구입 금액 ']

# 훈련 데이터와 테스트 데이터로 분할
x_train, x_test, y_train, y_test = sms.train_test_split(x, y, train_size=0.5, test_size=0.5)

import sklearn.linear_model as slm

lr_Model = slm.LinearRegression()
lr_Model.fit(x_train,y_train)
Train_score = lr_Model.score(x_train,y_train)
Test_score = lr_Model.score(x_test,y_test)

print('-----------')

print('다중 선형 회귀 - Train_score:',Train_score,'Test_score:',Test_score)
```

```python
import statsmodels.api as sm
# statsmodels를 사용하여 회귀 분석
x_train_sm = sm.add_constant(x_train)  # 상수항 추가
model = sm.OLS(y_train, x_train_sm).fit()

# 회귀 결과 출력
print(model.summary())
```

```python
# 다중공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor
import sklearn.model_selection as sms

# VIF 계산
vif = pd.DataFrame()
vif["변수"] = x_train_sm.columns
vif["VIF"] = [variance_inflation_factor(x_train_sm.values, i) for i in range(x_train_sm.shape[1])]

print(vif)
```

```python
# 잔차 분석
residuals = model.resid

# Q-Q plot
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot')
plt.show()

# 잔차 vs 예측값
plt.scatter(model.fittedvalues, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

```

```python
import pandas as pd
import sklearn.model_selection as sms
import sklearn.linear_model as slm
import sklearn.decomposition as decomposition
import statsmodels.api as sm
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt

# 데이터 읽어오기
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

pd_DataSet = pd.read_csv('ml_sample/한국주택금융공사_주택금융신용보증 현황_20190630.csv', encoding='euc-kr')

DF = pd.DataFrame(pd_DataSet)

# 결측값이 있는 행 삭제
New_DF = DF.dropna(how='any')

# 독립 변수와 종속 변수 설정
x = New_DF[['연도', '월', '구입 세대수(호)', '전세 세대수(호)', '전세 금액 ', '중도금 세대수(호)',
            '중도금 금액 ', '개량 금액 ', '사업자보증세대수(호)', '사업자보증금액 ']]
y = New_DF['구입 금액 ']

# 주성분 분석 (PCA)
pca = decomposition.PCA()
x_pca = pca.fit_transform(x)

# 설명된 분산 비율 확인
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Scree Plot 그리기
plt.figure()
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', label='Individual explained variance')
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='x', label='Cumulative explained variance')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.legend()
plt.show()

# 최적의 주성분 수 선택 (예: 95% 이상의 분산을 설명하는 주성분 수)
optimal_components = np.argmax(cumulative_explained_variance >= 0.95) + 1
print(f'Optimal number of components: {optimal_components}')

# 선택된 주성분 수로 PCA 수행
pca = decomposition.PCA(n_components=optimal_components)
x_pca = pca.fit_transform(x)

# PCA 변환 후 훈련 데이터와 테스트 데이터로 분할
x_train, x_test, y_train, y_test = sms.train_test_split(x_pca, y, train_size=0.5, test_size=0.5, random_state=42)

# 선형 회귀 모델 적합
lr_Model = slm.LinearRegression()
lr_Model.fit(x_train, y_train)
Train_score = lr_Model.score(x_train, y_train)
Test_score = lr_Model.score(x_test, y_test)

# 상수항 추가하여 statsmodels로 요약
x_train_sm = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train_sm).fit()

print('-----------')
print('다중 선형 회귀 - Train_score:', Train_score, 'Test_score:', Test_score)
print(model.summary())

# VIF 계산
vif = pd.DataFrame()
vif["변수"] = ['const'] + [f'PC{i+1}' for i in range(x_train.shape[1])]
vif["VIF"] = [variance_inflation_factor(x_train_sm, i) for i in range(x_train_sm.shape[1])]

print(vif)

# 주성분의 구성 요소 확인
components = pd.DataFrame(pca.components_, columns=x.columns, index=[f'PC{i+1}' for i in range(pca.n_components_)])
print(components)
```
